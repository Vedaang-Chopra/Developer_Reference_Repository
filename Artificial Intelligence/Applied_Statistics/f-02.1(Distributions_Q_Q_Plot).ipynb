```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
# Generating a normal Distribtuion with 1000 samples
std_normal =np.random.normal(loc=0, scale=1, size=1000)
```

```python
# To Get the Z Percentile:
Z=10
np.percentile(std_normal, Z)
```

```python
# Generating a distribution for evaluation...
measurement_variable=np.random.normal(loc=5, scale=10, size=100)
```

```python
#  Plotting a Q-Q-Plot
# Here the stats probplot compares the distribution, with normal(norm) distribution
stats.probplot(measurement_variable, dist="norm", plot=pylab)
print()
# Here X and Y are almost on same line, therefore they have similar distribution, that is Gaussian
```

```python
# Generating distributions for evaluation to understand effect of number of observations in a distribtion...
measurement_variable_1=np.random.normal(loc=5, scale=10, size=1000)
measurement_variable_2=np.random.normal(loc=5, scale=10, size=50)
measurement_variable_3=np.random.normal(loc=5, scale=10, size=50000)
measurement_variable_4=np.random.normal(loc=5, scale=10, size=10)
```

```python
stats.probplot(measurement_variable_1, dist="norm", plot=pylab)
print()
```

```python
stats.probplot(measurement_variable_2, dist="norm", plot=pylab)
print()
```

```python
stats.probplot(measurement_variable_3, dist="norm", plot=pylab)
print()
```

```python
stats.probplot(measurement_variable_4, dist="norm", plot=pylab)
print()
```

If number of observations are small, QQ Plot is harder to interpret(wrong results!!!)

## Changing Distributions

```python
#  Plotting a Q-Q-Plot
# Here the stats probplot compares the distribution, with uniform distribution
# Generating a unifotm distribution for evaluation...
measurement_variable=np.random.uniform(low=1, high=10, size=100)
stats.probplot(measurement_variable, dist="norm", plot=pylab)
print()
```
