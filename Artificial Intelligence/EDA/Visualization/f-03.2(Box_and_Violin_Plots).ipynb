```python
#  Creating a DataFrame.......
iris_data_complete=pd.DataFrame(dataset['data'],columns=[i.replace(" (cm)",'').strip().replace(' ','_') for i in dataset['feature_names']])
iris_data_complete['species_label']=dataset['target']
species=[]
for i in dataset['target']:
        species.append(dataset['target_names'][i])
iris_data_complete['species']=species
iris_data_complete.head()
```

### Box Plots
<ul>
<li> Another implementation of 1-D scatter plot</li>
<li>Gives information about the Quantiles, IQR, Median and outliers....</li>
<li> Uses IQR for whiskers</li>
<li> In the box plot, top line of the box is 75th Percentile, the middle line of the box is 50th Percentile and bottom line of the box is 25th percentile.
<li> The exetension on the box, are called whiskers and are equal to 1.5 IQR, and after that are all outliers</li>
<li> Whiksers can also be max and min value for the data given</li>
</ul>

```python
fig, ax = plt.subplots(2, 2, figsize=(30, 20))
sns.boxplot(data=iris_data_plot, y="petal_length", x="species", ax=ax[0,0])
sns.boxplot(data=iris_data_plot, y="sepal_length", x="species", ax=ax[0,1])
sns.boxplot(data=iris_data_plot, y="sepal_width", x="species",  ax=ax[1,0])
sns.boxplot(data=iris_data_plot, y="petal_width", x="species",  ax=ax[1,1])
plt.grid()
plt.show()
print()
```

### Violin Plots
It combines the advantages of box plot and distribution plot.
<ul>
<li> The violin plots has the box plot inside it.</li>
<li> The inner line is similar to box plots box. The white point is median(50th percentile), the hard line is the 25th and 75th quantile plot and then the whisker with light line</li>
<li> It also plots the distribution with the box</li>
</ul>

```python
fig, ax = plt.subplots(2, 2, figsize=(30, 20))
sns.violinplot(data=iris_data_plot, x="petal_length", y="species", ax=ax[0,0])
sns.violinplot(data=iris_data_plot, x="sepal_length", y="species", ax=ax[0,1])
sns.violinplot(data=iris_data_plot, x="sepal_width", y="species",  ax=ax[1,0])
sns.violinplot(data=iris_data_plot, x="petal_width", y="species",  ax=ax[1,1])
plt.grid(visible=True)
plt.show()
print()
```
