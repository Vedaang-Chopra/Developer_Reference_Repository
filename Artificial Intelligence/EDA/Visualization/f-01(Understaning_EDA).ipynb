<h5> <i> <u> Import Statements </u> </h5>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing Iris Dataset.........................
from sklearn.datasets import load_iris
dataset=load_iris()
```

<h3> <i> EDA Steps ................... </h3>

```python
dataset.keys()
```

```python
dataset['target_names']
```

```python
#  Creating a DataFrame.......
iris_data_complete=pd.DataFrame(dataset['data'],columns=[i.replace(" (cm)",'').strip().replace(' ','_') for i in dataset['feature_names']])
iris_data_complete['species_label']=dataset['target']
species=[]
for i in dataset['target']:
        species.append(dataset['target_names'][i])
iris_data_complete['species']=species
iris_data_complete.head()
```

```python
#  Complete Avaiable Dataset............
iris_data_complete.shape
```

```python
#  Check if dataset is balanced or not...................
iris_data_complete['species'].value_counts()
```

<h3> <i> Starting Basic Vizulisation................ </h3>

```python
#  Note in every plot, check the scale, x-axis and y-axis............
iris_data_complete.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
#  Plotting 2-D scatter plot based on dependent variable..............
sns.set_style("whitegrid")
sns.scatterplot(data=iris_data_complete, x='sepal_length', y='sepal_width', hue="species")
```

```python
#  Plotting 2-D scatter plot based on dependent variable-2..............
sns.set_style("whitegrid")
sns.FacetGrid(iris_data_complete,hue="species", size=4).map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()
plt.show()
```
