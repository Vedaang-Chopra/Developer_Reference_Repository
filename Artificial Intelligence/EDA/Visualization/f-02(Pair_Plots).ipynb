<h5> <i> <u> Import Statements </u> </h5>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing Iris Dataset.........................
from sklearn.datasets import load_iris
dataset=load_iris()
```

<h3> <i> EDA Steps ................... </h3>

```python
dataset.keys()
```

```python
dataset['target_names']
```

```python
#  Creating a DataFrame.......
iris_data_complete=pd.DataFrame(dataset['data'],columns=[i.replace(" (cm)",'').strip().replace(' ','_') for i in dataset['feature_names']])
iris_data_complete['species_label']=dataset['target']
species=[]
for i in dataset['target']:
        species.append(dataset['target_names'][i])
iris_data_complete['species']=species
iris_data_complete.head()
```

```python
#  Complete Avaiable Dataset............
iris_data_complete.shape
```

```python
#  Check if dataset is balanced or not...................
iris_data_complete['species'].value_counts()
```

<h3> <i> Understanding Pair Plots................ </h3>

<b> Pair Plots :</b> Pair of all features plotted   <br>
<ol>
<li>Number of pair Plots is: (No.of Features C 2) <br></li>
<li>Focus only on the upper triangle or lower triangle, since it is a reverse of each other</li>
<li> Pair Plot Disadvantages </li>
<ul>
<li> Can not be used with high number of features  </li>
<li> Generally Features upto 6, pair plots are managable </li>
<li> Cannot visualize higher dimensional patterns in 3-D and 4-D</li>
<li> Only used for 2-D Patterns</li>
</ul>
</ol>

```python
iris_data_plot=iris_data_complete.copy()
del iris_data_plot['species_label']
```

```python
sns.set_style("whitegrid")
sns.pairplot(iris_data_plot, hue="species", size=4)
plt.show()
```
