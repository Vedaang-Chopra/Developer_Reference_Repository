```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
```

```python
t1=np.arange(0.0,5,0.1)
t2=np.arange(0.0,5,0.02)
plt.figure(1, figsize=(20,10))
plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),color='b')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r')
plt.show()
```
