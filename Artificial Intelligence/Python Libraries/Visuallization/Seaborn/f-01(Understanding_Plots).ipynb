```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
dataset=pd.DataFrame(iris['data'],columns=iris['feature_names'])
dataset['target']=iris['target']
```

```python
dataset.columns
```

```python
# plt.figure()
parameters={
    'hue':'target',
    'figure':plt.figure(figsize=(20,10))
}
plt_parameters={
    # 'linewidth':20
    
}
sns.lineplot(data=dataset, x='sepal length (cm)', y='sepal width (cm)',**parameters)
```

```python
fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(30)
parameters={
    'hue':'target',
    # 'figure':plt.figure(figsize=(20,10))
}

sns.lineplot(data=dataset, x='sepal length (cm)', y='sepal width (cm)',**parameters,ax=ax, palette='GnBu', label='Sepal Length vs Sepal Width', legend=False)
sns.lineplot(data=dataset, x='sepal length (cm)', y='petal width (cm)',ax=ax,**parameters, label='Sepal Length vs Petal Length', legend=False)
plt.grid()
plt.legend(loc=0)
plt.show()
```

```python
# plt.figure()
parameters={
    'hue':'target',
    'figure':plt.figure(figsize=(20,10))
}
plt_parameters={
    # 'linewidth':20
    
}
sns.relplot(data=dataset, x='sepal length (cm)', y='sepal width (cm)',kind='line',col="petal width (cm)")
```
